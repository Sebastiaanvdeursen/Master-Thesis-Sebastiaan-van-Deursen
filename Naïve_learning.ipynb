```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

### With this code I do the numerical simulation for the tent map and upside-down tent map in combination with the na√Øve learning rule. It also gives multiple graphs which can be used to give information about the dynamics.

```python
# Define the asymmetric tent map
def tent_map_scalar(x, beta):
    if x <= (1 + beta) / 2 and x >= 0:
        return 2 * x / (1 + beta)
    elif x > (1 + beta) / 2 and x <= 1:
        return 2 * (1 - x) / (1 - beta)
    elif x < 0:
        return 0
    else:
        return 1

def tent_map_scalar_ud(x, beta):
    return 1 - tent_map_scalar(x, beta)

def naive_learning_forecast(x, t):
    return x[t - 2]

# Simulate the system under na√Øve learning
def simulate_naive_learning(beta, x0, tent, T=1000, burn_in=200):
    x_vals = np.zeros(T)
    forecast_vals = np.zeros(T + 1)
    x_vals[0] = x0
    if tent=="tent":
        for t in range(1, T):
            forecast_vals[t + 1] = naive_learning_forecast(x_vals, t + 1)
            x_vals[t] = tent_map_scalar(forecast_vals[t + 1], beta)
    elif tent=="tent_ud":
        for t in range(1, T):
            forecast_vals[t + 1] = naive_learning_forecast(x_vals, t + 1)
            x_vals[t] = tent_map_scalar_ud(forecast_vals[t + 1], beta)
    else:
        return False
    # Do not include the first 200 values
    return x_vals[burn_in:] 

def lyapunov_exponent_naive_learning(beta, x0, tent, T=5000, burn_in=1000, eps=1e-9):
    x_base = simulate_naive_learning(beta, x0, tent, T=T, burn_in=0)
    x_pert = simulate_naive_learning(beta, x0 + eps, tent, T=T, burn_in=0)
    dist = np.abs(x_pert - x_base) + 1e-20
    return np.log(dist[burn_in:] / eps).mean()

def orbit_and_lyapunov_learning_by_average(seed, n_skip, n_iter, tent, step=0.001):
    beta_range = np.linspace(-1, 1, int(1/step))
    lyap = []
    fig1, ax1 = plt.subplots(figsize=(8, 4))
    for beta in beta_range:
        x_vals = simulate_naive_learning(beta, seed, tent, T=n_iter + n_skip, burn_in=n_skip)
        ax1.scatter(np.full_like(x_vals, beta), x_vals, s=0.75, color='blue', alpha=0.8, linewidths=0)
        lyap.append(lyapunov_exponent_naive_learning(beta, seed, tent))
    ax1.set_xlim(-1, 1)
    ax1.set_ylim(0, 1)
    ax1.set_xlabel('beta')
    ax1.set_ylabel('x')
    plt.tight_layout()
    if tent=="tent":
        plt.savefig("Images/naive_learning_orbit_diagram.png", dpi=300)
    elif tent=="tent_ud":
        plt.savefig("Images/naive_learning_orbit_diagram_ud.png", dpi=300)
    plt.show()
    fig2, ax2 = plt.subplots(figsize=(8, 4))
    ax2.plot(beta_range, lyap, color='red')
    ax2.set_xlim(-1, 1)
    ax2.set_xlabel('beta')
    ax2.set_ylabel('lyapunov exponent')
    plt.tight_layout()
    if tent=="tent":
        plt.savefig("Images/naive_learning_lyapunov.png", dpi=300)
    elif tent=="tent_ud":
        plt.savefig("Images/naive_learning_lyapunov_ud.png", dpi=300)
    plt.show()

orbit_and_lyapunov_learning_by_average(0.4, 5000, 100, "tent", step=0.001)
```

```python
x = simulate_naive_learning(beta=0.6, x0=0.2, tent="tent", T = 200, burn_in=0)

def plot_time_series(x, tent):
    plt.figure(figsize=(8, 4))
    plt.plot(range(len(x)), x, color='black', linewidth=1)
    plt.xlabel("$t$", fontsize=12)
    plt.ylabel("$x_t$", fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    if tent=="tent":
        plt.savefig("Images/naive_learning_time_series.png")
    elif tent=="tent_ud":
        plt.savefig("Images/naive_learning_time_series_ud.png")
    else:
        return False
    plt.show()

plot_time_series(x, "tent")
```

```python
def tent_map(x, beta):
    x_arr = np.asarray(x)
    thresh = (1 + beta) / 2
    below = 2 * x_arr / (1 + beta)
    above = 2 * (1 - x_arr) / (1 - beta)
    return np.where(x_arr <= thresh, below, above)

def tent_map_ud(x, beta):
    return 1 - tent_map(x, beta)

dpi = 72

def plot_cobweb_naive_learning(beta, x0, tent="tent", n_iters=40,
                               fname=None, figsize=(600, 450)):
    m_vals = [x0]
    x_vals = []
    map_func = tent_map_scalar if tent == "tent" else tent_map_scalar_ud

    for _ in range(n_iters):
        x_next = map_func(m_vals[-1], beta)
        x_vals.append(x_next)
        m_vals.append(x_next)

    px, py = [m_vals[0]], [0] 
    for i in range(n_iters):
        # Vertical to the map
        px.append(m_vals[i])
        py.append(x_vals[i])
        # Horizontal to the next forecast
        px.append(m_vals[i+1])
        py.append(x_vals[i])

    x_plot = np.linspace(0, 1, 500)

    fig = plt.figure(figsize=(figsize[0] / dpi, figsize[1] / dpi), dpi=dpi)
    ax = fig.add_subplot(111)

    if tent=="tent":
        ax.plot(x_plot, tent_map(x_plot, beta),  c='#444444', lw=2,
                label=r'$T_{\beta, [0, 1]}(x)$', linestyle='--')
    elif tent=="tent_ud":
        ax.plot(x_plot, tent_map_ud(x_plot, beta),  c='#444444', lw=2,
                label=r'$\hat{T}_{\beta, [0, 1]}(x)$', linestyle='--')
    else:
        return False
    ax.plot(x_plot, x_plot, c='#444444', lw=2, label=r'$y=x$')

    ax.plot(px, py, c='b', alpha=0.75, label='trajectory')

    ax.set_aspect('equal')
    ax.minorticks_on()
    ax.grid(which='both', alpha=0.5)
    ax.set_xlabel(r'$x_t$')
    ax.set_ylabel(r'$x_{t + 1}$')
    ax.legend(loc='best')

    if tent=="tent":
        plt.savefig("Images/naive_learning_cobweb.png")
    elif tent=="tent_ud":
        plt.savefig("Images/naive_learning_cobweb_ud.png")
    else:
        return False

beta = 0.60
x0 = 0.20
plot_cobweb_naive_learning(beta, x0, tent="tent", n_iters=10)
```
